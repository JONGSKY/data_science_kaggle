```python
def get_necessary_df(x_train_df, y_train_df):
    all_df = x_train_df.join(y_train_df)
    corr = all_df.corr(method = 'pearson')
    necessary_df_index = pd.DataFrame(corr[(corr['date']<=-0.2) | (corr['date']>=0.2)]['date']).index
    necessary_df = all_df[necessary_df_index]
    necessary_df = necessary_df.drop(['date'], axis=1)
    return necessary_df
```
